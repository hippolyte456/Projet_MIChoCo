### Variables catégorielles binaires

Elles seront par la suite codées en 0 et 1.

````python
#on regarde quelles sont les variables binaires
print("Genre : ", df["Genre"].unique())        #Genre dans fichiers d'Octobre
print("Régime : ", df["Régime"].unique())        #Régime particulier
print("Connaissance entourage : ", df['Connaissance entourage'].unique())        #Connaissance
print("Influence aspect : ", df['Influence aspect'].unique())
print("Influence entourage : ", df['Influence entourage'].unique())        #Non binaire mas Influence sociale
````

Pour `Influence aspect` on voit un espace après le 'non' qui est comptabilisé comme une valeur différente. Il faut le modifier. Pour le `Genre` il y a des valeurs 'NB' (Non binaire ?) comment les traiter ? Pour le moment on les laisse en NA.

````python
df["Genre"] = df["Genre"].replace({'Homme':1, 'Femme':0, 'NB':np.nan})
df["Régime"] = df["Régime"].replace({'non':0, 'oui':1})
df['Influence aspect'] = df['Influence aspect'].replace(' ', '', regex=True)
df['Influence aspect'] = df['Influence aspect'].replace({'non':0, 'oui':1})
df['Influence entourage'] = df['Influence entourage'].replace({'non':1, 'oui':2}) #our correspondre à la variable échelle d'Octobre
df['Connaissance entourage'] = df['Connaissance entourage'].replace({'non':0, 'oui':1})
````

````python
print("Genre : ", df["Genre"].unique())       
print("Régime : ", df["Régime"].unique())      
print("Connaissance entourage : ", df['Connaissance entourage'].unique())     
print("Influence aspect : ", df['Influence aspect'].unique())
print("Influence entourage : ", df['Influence entourage'].unique()) 
````

### Variables categorielles en échelle

````python
#on regarde quelles sont les variables concernées
print("Fréquentation : ", df['Fréquentation'].unique())        #Fréquence CROUS
print("Faim : ", df['Faim'].unique())        #Faim
print("Quantité : ", df['Quantité'].unique())        
print("Equilibre : ", df['Equilibre'].unique())        #Equilibre alim
print("Aspect : ", df['Aspect'].unique())        #Perception senso
print("Attente : ", df['Attente'].unique())        #Nb file
````

#### Fréquentation

````python
#on remplace les valeurs de la colonne par des valeurs numériques
df["Fréquentation"] = df["Fréquentation"].replace('<1', 0)
df["Fréquentation"] = df["Fréquentation"].replace('1à3', 1)
df["Fréquentation"] = df["Fréquentation"].replace('>1', 1)
df["Fréquentation"] = df["Fréquentation"].replace('>3', 2)
````

````python
print("Fréquentation : ", df['Fréquentation'].unique())
````

#### Faim

````python
#Changement d'échelle de 0 à 3
df["Faim"] = df["Faim"].replace([1, 2, 3],0)
df["Faim"] = df["Faim"].replace([4, 5],1)
df["Faim"] = df["Faim"].replace([6, 7],2)
df["Faim"] = df["Faim"].replace([8, 9, 10],3)
````

````python
print("Faim : ", df['Faim'].value_counts())  
````

#### Equilibre

Dans un premier temps on règle le problème du 'è'. On suppose qu'il s'agit d'un '7' mal tapé sur un clavier AZERTY. 

````python
df["Equilibre"] = df["Equilibre"].replace('è','7')
````

Nous changeons l'échelle pour qu'elle soit la même que celle utilisée dans les formulaires d'Octobre. Pour ce faire nous changeons au préalable le type de données de chaines de caractères à entiers.

````python
#Changement du type
df["Equilibre"] = df["Equilibre"].astype(int)

#Changement d'échelle 0 à 4.
df["Equilibre"] = df["Equilibre"].replace([0, 1, 2],0)
df["Equilibre"] = df["Equilibre"].replace([3, 4],1)
df["Equilibre"] = df["Equilibre"].replace([5, 6],2)
df["Equilibre"] = df["Equilibre"].replace([7, 8],3)
df["Equilibre"] = df["Equilibre"].replace([9, 10],4)
````

````python
print("Attente : ", df['Attente'].value_counts()) 
````

#### Attente

````python
#Changement d'échelle de 0 à 2
df["Attente"] = df["Attente"].replace([0, 1, 2, 3],0)
df["Attente"] = df["Attente"].replace([4, 5, 6],1)
df["Attente"] = df["Attente"].replace([7, 8, 9, 10],2)
````

````python
print("Attente : ", df['Attente'].value_counts())
````

#### Aspect

````python
#Changement d'échelle de 0 à 3
df["Aspect"] = df["Aspect"].replace([1, 2, 3],0)
df["Aspect"] = df["Aspect"].replace([4, 5],1)
df["Aspect"] = df["Aspect"].replace([6, 7],2)
df["Aspect"] = df["Aspect"].replace([8, 9, 10],3)
````

````python
print("Aspect : ", df['Aspect'].value_counts())
````
