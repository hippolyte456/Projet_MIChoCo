### Variables catégorielles binaires

Elles seront par la suite codées en 0 et 1.

````python
#on regarde quelles sont les variables binaires
print("Genre : ", df["Genre"].unique())        #Genre dans fichiers d'Octobre
print("Regime : ", df["Regime"].unique())        #Régime particulier
print("Connaissances : ", df['Connaissances'].unique())        #Connaissance
print("Influence aspect : ", df['Influence aspect'].unique())
print("Influence : ", df['Influence'].unique())        #Non binaire mas Influence sociale
````

Pour `Influence aspect` on voit un espace après le 'non' qui est comptabilisé comme une valeur différente. Il faut le modifier. Pour le `Genre` il y a des valeurs 'NB' (Non binaire ?) comment les traiter ? Pour le moment on les laisse en NA.

````python
df["Genre"] = df["Genre"].replace({'Homme':1, 'Femme':0, 'NB':np.nan})
df["Regime"] = df["Regime"].replace({'non':0, 'oui':1})
df['Influence aspect'] = df['Influence aspect'].replace('7', np.nan, regex=True)
df['Influence aspect'] = df['Influence aspect'].replace({'non':0, 'oui':1})
df['Influence'] = df['Influence'].replace({'non':1, 'oui':2}) #our correspondre à la variable échelle d'Octobre
df['Connaissances'] = df['Connaissances'].replace({'non':0, 'oui':1})
````

````python
print("Genre : ", df["Genre"].unique())        
print("Regime : ", df["Regime"].unique())        
print("Connaissances : ", df['Connaissances'].unique())        
print("Influence aspect : ", df['Influence aspect'].unique())
print("Influence : ", df['Influence'].unique())        
````
