```python
def round_to_nearest_IMC_float(x) : 
    if math.isnan(x):
        return x
    
    modulo = round(x)%5
    #print(x, modulo)
    if modulo > 2.5 :
        to_return = round(x) + (5-modulo)
        return to_return
    
    elif modulo == 0 : 
        return round(x)
    
    else : 
        to_return = round(x) - modulo
        return to_return
    
df["IMC"] = df["IMC"].map(lambda IMC : round_to_nearest_IMC_float(IMC))
print(df["IMC"])
```
