## Problème des Numéro de formulaires communs Juin ENS et LDV

```python
print(juin_ens['Numéro de formulaire'].unique())
print(juin_ldv['Numéro de formulaire'].unique())
```

On remarque qu'ils sont numérotés de la même façon donc que 'B11' par exemple est dans les deux dataframes ce qu'on ne souhaite pas. Il faut donc corriger la colonne `Numéro de formulaire` mais aussi les colonnes `Ami`.

```python
for col_num_form in ['Numéro de formulaire', 'Ami 1', 'Ami 2', 'Ami 3', 'Ami 4', 'Ami 5', 'Ami 6',
       'Ami 7', 'Ami 8', 'Ami 9', 'Ami 10', 'Ami 11', 'Ami 12', 'Ami 13',
       'Ami 14', 'Ami 15', 'Ami 16', 'Ami 17']:
    juin_ldv[col_num_form] = juin_ldv[col_num_form].replace(to_replace=r'^A', value='E', regex=True)
    juin_ldv[col_num_form] = juin_ldv[col_num_form].replace(to_replace=r'^B', value='F', regex=True)
```

## Preparation du merged

```python
octobre_all
```

```python
#list(set(juin_ldv.columns + juin_ens.columns))

final = octobre_all.copy()
entrees = ['saladesoja','carottesrapees', 'saladeriz', 'saucisson', 'endives', 'sardines'] + ['oeufmimosa', 
           'betteravespommesechalotes', 'tomates', 'pizzaentree','saladepommesdeterre']
plats = ['pavedecabillaud', 'lasagnesvg', 'chipo', 'pizza', 'burger', 'semoule',
       'frites'] + ['escalopedeporc', 'grillades', 'tartesaumon', 'gnocchispestovege', 'pommesdeterrevapeur',
       'poeleemeridionale']
desserts = ['fruit', 'yaourtnature', 'donut', 'crèmedessert', 'emmental',
       'tartefrboisesananas', 'moussechocolat', 'chaource'] + ['tartechocolat', 'madeleine',
       'fromage', 'pecheausirop']


for plat in entrees :
    if plat not in final.columns:
        final.insert(int(np.where(final.columns=='torsades')[0]), plat, np.nan)
        
for plat in plats :
    if plat not in final.columns:
        final.insert(int(np.where(final.columns=='yaourtauxfruits')[0]), plat, np.nan)
        
for plat in desserts :
    if plat not in final.columns:
        final.insert(len(final.columns), plat, np.nan)
```

```python
csv_tot = pd.concat([final, juin_ens, juin_ldv])
csv_tot = csv_tot.set_index("Numéro de formulaire")
csv_tot = csv_tot.drop('Unnamed: 0', axis=1)
print(csv_tot.shape)
csv_tot.head()
```
