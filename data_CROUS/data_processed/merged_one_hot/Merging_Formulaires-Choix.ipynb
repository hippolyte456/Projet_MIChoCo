## Sauvegarde d'un fichier concatenant formulaire et choix pour chaque sujet

```python
#concatenation du formulaire avec les choix 
choix = pd.read_csv("Merged_Choix.csv", encoding='latin1', sep=",") 
formulaire = pd.read_csv("Merged_Formulaires.csv", encoding='latin1')
```

```python
formulaire
```

```python
choix
```

#### Merging

```python
#on associe choix et formulaire par le numero de formulaire
choix = choix.set_index("Numéro de formulaire ")
formulaire = formulaire.set_index("N° Formulaire")
#on merge les deux dataframes
final = pd.merge(formulaire, choix, left_index=True, right_index=True)
```

#### Suppression des lignes en double

```python
final.shape
```

```python
final.reset_index(inplace = True) 
```

```python
final.drop_duplicates(subset = 'index', keep = 'first', inplace=True)
```
