```python
cpt = 112
for onehot_col in df.columns[118:] :
    count_col = np.array(df[df.columns[cpt]])
    df[onehot_col] = np.where(count_col > 0,1,0)
    cpt+=1
df
```

```python
#Réduction en classes des colonnes avec des valeurs très diverses pour allégre le one-hot encoding
def round_to_nearest_IMC(x) : 
    modulo = int(x.split(",")[0])%5
    print("Data : ", int(x.split(",")[0]), modulo)
    if modulo > 2.5 :
        to_return = int(x.split(",")[0]) + (5-modulo)
        print("Return : ", to_return)
        return to_return
    
    elif modulo == 0 : 
        print("Return : ", int(x.split(",")[0]))
        return int(x.split(",")[0])
    
    else : 
        to_return = int(x.split(",")[0]) - modulo
        print("Return : ", to_return)
        return to_return
    
df["IMC"] = df["IMC"].map(lambda IMC : round_to_nearest_IMC(IMC))
print(df["IMC"])
```
